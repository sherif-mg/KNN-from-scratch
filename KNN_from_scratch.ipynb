```python
import pandas as pd
from sklearn.utils import shuffle
import numpy as np
from collections import Counter
from sklearn.metrics import mean_squared_error
```

```python
def KNN(k,trainData,labels,point,classification=True):
  distance=np.sum(np.square(trainData-point),axis=1)
  trainIndexes=np.argsort(distance,axis=-1)
  trainData=trainData[trainIndexes]
  labels=labels[trainIndexes]
  if(classification):
    return Counter(labels[0:k]).most_common()[0][0]
  return np.average(labels[0:k])
```
